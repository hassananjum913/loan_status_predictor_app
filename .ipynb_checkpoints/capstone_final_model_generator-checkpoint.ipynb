```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn  as sns
import sklearn
import os
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
# clean_bal_df=pd.read_csv(r'balanced_cleaned_train_data.csv')
# clean_bal_df.head()
```

```python
clean_bal_df=pd.read_csv(r'credit_train_cleaned.csv')
clean_bal_df.head()
```

```python
clean_bal_df= pd.get_dummies(clean_bal_df,drop_first=True)
```

```python
clean_bal_df.head()
```

```python
clean_bal_df.columns
```

```python
imp_features = ['Annual Income','Home Ownership_Home Mortgage','Credit Score',
                'Current Loan Amount','Home Ownership_Rent','Home Ownership_Own Home',
                'Term_Short Term','Years in current job_10+ years',
                'Purpose_Debt Consolidation','Monthly Debt',
                'Current Credit Balance','Number of Credit Problems',
                'Maximum Open Credit','Years in current job_2 years',
                'Years in current job_3 years','Number of Open Accounts',
                'Years of Credit History','Years in current job_< 1 year',
                'Months since last delinquent','Years in current job_4 years']
```

```python
# splitting data
x=clean_bal_df[imp_features]
y=clean_bal_df['Loan Status_Fully Paid']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x.head(10)
```

```python
y.head(50)
```

### Model: 
* Hyper Paramter Tuned RandomForestClassifier with Important Features.

```python

#  model fitting


random_forest_clf =  RandomForestClassifier(random_state=10,n_estimators=400,min_samples_split=5,
                                     max_features='auto',max_depth=60,bootstrap=False)

random_forest_clf.fit(x_train, y_train)

y_pred_imp_feature = random_forest_clf.predict(x_test)
```

```python
from sklearn.model_selection import cross_val_score
```

### 5-Fold Cross Validation

```python
cross_val_score(estimator=random_forest_clf,X=x_test,y=y_test,cv=5).mean()
```

```python
from sklearn.metrics import confusion_matrix
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(confusion_matrix(y_pred_imp_feature,y_test),annot=True,fmt='d',cmap='Blues',cbar=False)
plt.xlabel('Predicted Labels')
plt.ylabel('Actual Labels')
plt.show()
```

```python
import pickle

file=open('loan_status_prediction_model.pkl','wb')

pickle.dump(random_forest_clf,file)
file.close()
```
